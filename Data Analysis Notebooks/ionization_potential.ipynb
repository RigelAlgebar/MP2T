```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('ticks')
sns.set_context('paper')
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
```

<h1 align="center">Nucleobases</h1> 

```python
ipdf_1 = pd.read_excel('EOM-MP2T.xlsx',sheet_name='set1_IP_nucleobases', skiprows=1)
ipdf_1.head()
```

```python
#Dataframe for normal errors

err_ip_df_1 = ipdf_1[['EOM-MP2.1', 'EOM-MP2T.1']] #We take the specific columns
err_ip_df_1.columns = ['MP2', 'MP2T'] # Now just rename the columns
err_ip_df_1 = err_ip_df_1.dropna() #We ignore the non-values encountered

err_ip_df_1.head() #This juts show the first values of the columns to make sure that we took the right ones
```

```python
#Now just a quick statistical description of the datasets

err_ip_df_1.describe()
```

```python
#Dataframe for absolute errors

abs_ip_df_1 = ipdf_1[['EOM-MP2.2', 'EOM-MP2T.2']] #We take the specific columns
abs_ip_df_1.columns = ['Abs MP2', 'Abs MP2T'] # Now just rename the columns
abs_ip_df_1 = abs_ip_df_1.dropna() #We ignore the non-values encountered


abs_ip_df_1.head() #This juts show the first values of the columns to make sure that we took the right ones
```

```python
#Now just a quick statistical description of the datasets

abs_ip_df_1.describe()
```

```python
#Define the bins for the histograms
#Remember, if you use the key word "auto" in the bins parameter, Seaborn will fit the optimal number

bins_err_1 = np.linspace(0, 0.4, 8)
bins_abs_1 = np.linspace(0, 0.4, 8)
```

```python
#Define plot size in inches (width, height) & resolution(DPI) and put a white background
fig = plt.figure(figsize=(18, 12), dpi=100)
sns.set_style("white")


#Define the data
hist1 = err_ip_df_1['MP2']
hist2 = err_ip_df_1['MP2T']

#plot data
plt.hist(hist1, bins_err_1, histtype='stepfilled', label='MP2', color='blue')
plt.hist(hist2, bins_err_1, histtype='stepfilled', alpha=0.5, label='MP2T', color='red')

#configure  X axesgure(figsize=(15, 10)


#configure  Y axes


#labels
plt.xlabel("Err (Ev)", size=25)
plt.ylabel("Frequency", size=25)

#define font size
sns.set(font_scale=2.5)

#title
plt.title("Nucleobases IP")



#define font size
sns.set(font_scale=2.5)


plt.legend(loc='upper right')

plt.legend(loc='upper right')
plt.show()
```

```python
#Define plot size in inches (width, height) & resolution(DPI) and put a white background
fig = plt.figure(figsize=(18, 12), dpi=100)
sns.set_style("white")


#Define the data
hist3 = abs_ip_df_1['Abs MP2']
hist4 = abs_ip_df_1['Abs MP2T']

#plot data
plt.hist(hist3, bins_abs_1, histtype='stepfilled', label='MP2', color='blue')
plt.hist(hist4, bins_abs_1, histtype='stepfilled', alpha=0.5, label='MP2T', color='red')

#configure  X axes


#configure  Y axes


#labels
plt.xlabel("Err (Ev)", size=25)
plt.ylabel("Frequency", size=25)

#define font size 592 
sns.set(font_scale=2.5)

#title
plt.title("Nucleobases IP Abs")

plt.axvline(0.1, color='black', linestyle='dashed')
plt.legend(loc='upper right')
plt.show()
```

```python
sns.set_style("white")
fig = plt.figure(figsize=(15, 10))
c1, c2 = sns.color_palette("seismic", 2)


#define font size
sns.set(font_scale=2.5)

#title
plt.title("Nucleobases IP")

plt.ylim(0,10.5)

sns.kdeplot(err_ip_df_1.MP2, kernel='gau', label='MP2', shade=True, color=c1)
sns.kdeplot(err_ip_df_1.MP2T, kernel='gau',  label='MP2T',shade=True, color=c2)

plt.axvline(0, color='black', linestyle='dashed')
plt.show()
```

```python
sns.set_style("whitegrid")
fig = plt.figure(figsize=(15, 10))
plt.axhline(0, color='black')

#define font size
sns.set(font_scale=2.5)

#title
plt.title("Nucleobases IP")


# Draw a nested violinplot and split the violins for easier comparison
viol = sns.violinplot(data=err_ip_df_1, split=True, inner='quartile', palette='seismic')
```
